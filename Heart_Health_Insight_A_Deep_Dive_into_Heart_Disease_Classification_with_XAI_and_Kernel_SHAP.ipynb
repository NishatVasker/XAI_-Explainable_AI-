# **X-AI_D SCORE**

```python
def xai_d_score(y_true, y_pred):
    """
    XAI-D Score

    This score takes into account the class prediction of the sentence as well as the fidelity of the explanation provided.
    The score is a function of the area under the ROC curve (AUC-ROC) and the F1-score of the classifier.
    """
    from sklearn.metrics import roc_auc_score, f1_score

    # Calculate the area under the ROC curve (AUC-ROC)
    auc_roc = roc_auc_score(y_true, y_pred, average='weighted', multi_class='ovr')

    # Calculate the F1-score of the classifier
    f1 = f1_score(y_true, y_pred, average='weighted')

    # Combine the AUC-ROC and F1-score into a single XAI-D score
    xai_d_score = auc_roc * f1

    return xai_d_score
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Create partial dependence plots for selected features
selected_features = ['age', 'chol', 'thalach', 'trestbps']

# Set the grid resolution
grid_resolution = 50

# Create a grid of subplots
n_rows = len(selected_features)
n_cols = 2  # Adjust the number of columns as needed
fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 2 * n_rows))
fig.subplots_adjust(wspace=0.4, hspace=0.4)

for i, feature in enumerate(selected_features):
    ax = axes[i // n_cols, i % n_cols]

    # Define the values for the feature
    feature_values = X_test_norm[:, X_test.columns.get_loc(feature)]

    # Create a range of values for the partial dependence plot
    values = np.linspace(min(feature_values), max(feature_values), grid_resolution)

    # Initialize an array to store partial dependence values
    pdp_values = np.zeros((grid_resolution,))

    # Calculate partial dependence values
    for j, value in enumerate(values):
        instance_copy = X_test_norm.copy()
        instance_copy[:, X_test.columns.get_loc(feature)] = value
        pdp_values[j] = classifier.predict_proba(instance_copy)[:, 1].mean()

    # Plot the partial dependence plot with custom styling
    sns.lineplot(x=values, y=pdp_values, ax=ax, color='red', linewidth=3)
    ax.set_title(f'Partial Dependence Plot for {feature}')
    ax.set_xlabel(feature)
    ax.set_ylabel('Partial Dependence')
    ax.grid(True)

# Remove any unused subplots
for i in range(len(selected_features), n_rows * n_cols):
    fig.delaxes(axes[i // n_cols, i % n_cols])

plt.show()
```
